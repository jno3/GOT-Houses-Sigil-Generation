```python
import pandas as pd
import numpy as np
import urllib.request
import requests
from bs4 import BeautifulSoup
import os
from PIL import Image
from cairosvg import svg2png
```

```python
df = pd.read_csv('gameofthrones.csv')
useful = df.pop('Sigil')
```

```python
useful.isnull().unique()
```

```python
path = './Images'
try: 
    os.mkdir(path)
except OSError as error: 
    print(error)

unique = []    

for i, sigil in enumerate(useful):
    file_name = df['House'].iloc[i]
    html_url = sigil
    file_path = path + '/' + file_name + '.png'
    file_path2 = path + '/' + file_name + '.jpg'

    html = requests.get(html_url).content
    soup = BeautifulSoup(html, 'html.parser')
    url_2 = soup.find("a", {"class": "internal"}).get('href')
    if('.png' not in url_2):
        url = 'https://awoiaf.westeros.org' + url_2
    
    img_data = requests.get(url).content
    
    #print(url)
    if(img_data not in unique):
        unique.append(img_data)
        
        # parser = etree.XMLParser(recover=True)
        # img_data = etree.fromstring(img_data, parser=parser)
        
        svg2png(bytestring=img_data, write_to=file_path)
        
        im = Image.open(file_path)
        new_image = im.resize((128, 128))
            # new_width, height = im.size
            # new_height = height+2*30
            # result = Image.new(im.mode, (new_width, new_height), (0, 0, 0, 0))
            # result.paste(im)
            #print(file_path)
        new_image = new_image.convert('RGB')
        new_image.save((file_path2))
```
